```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'Accuracy Score': [0, 0.5, 1, 1.5, 2],
    'Baseline': [44, 48, 26, 8, 3],
    'Stepwise': [17, 61, 41, 10, 3],
    'Visual-First': [18, 45, 46, 17, 3],
    'Explanation-First': [35, 42, 27, 20, 5]
}

df = pd.DataFrame(data)
df_melted = df.melt(id_vars='Accuracy Score', var_name='CoT Structure', value_name='Count')


sns.set(style='whitegrid', font_scale=1.2)


plt.figure(figsize=(10, 6))
sns.barplot(x='Accuracy Score', y='Count', hue='CoT Structure', data=df_melted, palette='Set2')

plt.title('Accuracy Score Distribution Across CoT Structures')
plt.xlabel('Accuracy Score')
plt.ylabel('Count')

plt.legend(title='CoT Structure', loc='upper right')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Accuracy Score': [0, 0.5, 1, 1.5, 2],
    'Baseline': [44, 48, 26, 8, 3],
    'Stepwise': [17, 61, 41, 10, 3],
    'Visual-First': [18, 45, 46, 17, 3],
    'Explanation-First': [35, 42, 27, 20, 5]
}

df = pd.DataFrame(data)
df.set_index('Accuracy Score', inplace=True)

fig, axs = plt.subplots(1, 5, figsize=(20, 4))

colors = plt.get_cmap('Set2').colors

for idx, (score, row) in enumerate(df.iterrows()):
    axs[idx].pie(row, labels=row.index, autopct='%1.1f%%', colors=colors, startangle=90, counterclock=False)
    axs[idx].set_title(f'Accuracy = {score}')

plt.suptitle('Pie Charts for Each Accuracy Score', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

```python
import matplotlib.pyplot as plt

accuracy_scores = ['0', '0.5', '1', '1.5', '2']
baseline = [44, 48, 26, 8, 3]
stepwise = [17, 61, 41, 10, 3]
visual_first = [18, 45, 46, 17, 3]
explanation_first = [35, 42, 27, 20, 5]

cot_data = {
    'Baseline': baseline,
    'Stepwise': stepwise,
    'Visual-First': visual_first,
    'Explanation-First': explanation_first
}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854']

for i, (cot, counts) in enumerate(cot_data.items()):
    axes[i].pie(
        counts,
        labels=accuracy_scores,
        autopct='%1.1f%%',
        startangle=140,
        colors=colors,
        wedgeprops={'edgecolor': 'white'}
    )
    axes[i].set_title(f'{cot}')

plt.suptitle('Accuracy Score Distribution Within Each CoT Structure', fontsize=16)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['0', '0.5', '1', '1.5', '2']
baseline = [29, 13, 9, 4, 0]
stepwise = [8, 26, 15, 4, 2]
visual = [9, 21, 18, 6, 1]
explan = [28, 13, 5, 8, 1]

x = np.arange(len(labels))
width = 0.2


fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - 1.5*width, baseline, width, label='Baseline')
bars2 = ax.bar(x - 0.5*width, stepwise, width, label='Stepwise')
bars3 = ax.bar(x + 0.5*width, visual, width, label='Visual-First')
bars4 = ax.bar(x + 1.5*width, explan, width, label='Explanation-First')


ax.set_xlabel('Accuracy')
ax.set_ylabel('Count')
ax.set_title('LLaVA-Med Accuracy Score Distribution')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(bars1, padding=2, fontsize = 10)
ax.bar_label(bars2, padding=2, fontsize = 10)
ax.bar_label(bars3, padding=2, fontsize = 10)
ax.bar_label(bars4, padding=2, fontsize = 10)

plt.tight_layout()
plt.show()
```
